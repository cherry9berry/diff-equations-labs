```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')

# Параметры аппроксимации
terms = [1, 3, 10, 50]
colors = ['r', 'g', 'b', 'm']
x = np.linspace(0, 4 * np.pi, 1000)
```

```python
def square_wave(x):
    return 1.0 * (x % (2 * np.pi) < np.pi) - 1.0 * (x % (2 * np.pi) >= np.pi)

def fourier_series_square_wave(x, n_terms):
    y = np.zeros_like(x)
    for n in range(1, n_terms + 1):
        if n % 2 != 0:
            bn = 4 / (np.pi * n)
            y += bn * np.sin(n * x)
    return y
```

```python
y_exact = square_wave(x)

plt.figure(figsize=(12, 6))
plt.plot(x, y_exact, 'k-', linewidth=3, label='Оригинальная функция (Square Wave)')

for term, color in zip(terms, colors):
    y_approx = fourier_series_square_wave(x, term)
    plt.plot(x, y_approx, color=color, alpha=0.8, label=f'Сумма Фурье (N={term})')

plt.title('Аппроксимация функции рядом Фурье', fontsize=14)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
