# Подробный разбор решения Задачи №7 (Гиперболическое уравнение с граничными условиями третьего рода)

## Постановка задачи

**Дано:**
- **Уравнение:** $u_{tt} = u_{xx} + f(x,t)$, где $f(x,t) = 2\sin\frac{3\pi x}{2} + xe^{-t}$, $x \in [0,1]$, $t \geq 0$
- **Граничные условия третьего рода:**
  - При $x=0$: $\alpha u(0,t) - \beta u_x(0,t) = g_1(t)$, где $\alpha = 0$, $\beta = 1$ → $-u_x(0,t) = -\sin t$ → $u_x(0,t) = \sin t$
  - При $x=1$: $\gamma u(1,t) + \delta u_x(1,t) = g_2(t)$, где $\gamma = 1$, $\delta = 1$ → $u(1,t) + u_x(1,t) = 2\cos t + e^t$
- **Начальные условия:**
  - $u(x,0) = \sin\frac{3\pi x}{2} + x + 1$ (начальное смещение)
  - $u_t(x,0) = -\sin\frac{5\pi x}{2} - x + 1$ (начальная скорость)

**Особенность:** Это самая сложная задача — граничные условия третьего рода (комбинация функции и её производной). Собственные значения находятся из трансцендентного уравнения.


## Шаг 1: Приведение граничных условий к однородным

### Зачем это нужно?

Граничные условия **неоднородные третьего рода**, поэтому нужно найти функцию $s(x,t)$, которая "возьмет на себя" эти неоднородности.

### Поиск функции $s(x,t)$

Представим $u(x,t) = v(x,t) + s(x,t)$, где $s(x,t)$ должна удовлетворять:
- $s_x(0,t) = \sin t$
- $s(1,t) + s_x(1,t) = 2\cos t + e^t$

### Подбор функции $s(x,t)$

Пробуем найти $s(x,t)$ в виде квадратичной функции по $x$:
$$s(x,t) = a(t)x + b(t)x^2$$

где $a(t)$ и $b(t)$ — функции времени.

**Вычисляем производные:**
$$s_x(x,t) = a(t) + 2b(t)x$$
$$s(x,t) + s_x(x,t) = a(t)x + b(t)x^2 + a(t) + 2b(t)x = a(t) + (a(t) + 2b(t))x + b(t)x^2$$

**Применяем граничные условия:**

Из первого условия: $s_x(0,t) = a(t) = \sin t$, следовательно $a(t) = \sin t$

Из второго условия: $s(1,t) + s_x(1,t) = a(t) + (a(t) + 2b(t)) \cdot 1 + b(t) \cdot 1^2 = 2a(t) + 3b(t) = 2\cos t + e^t$

Подставляем $a(t) = \sin t$:
$$2\sin t + 3b(t) = 2\cos t + e^t$$
$$3b(t) = 2\cos t + e^t - 2\sin t$$
$$b(t) = \frac{2\cos t + e^t - 2\sin t}{3}$$

### Результат

$$s(x,t) = x\sin t + \frac{x^2}{3}(2\cos t + e^t - 2\sin t)$$

**Итог шага 1:** Функция $s(x,t)$ удовлетворяет граничным условиям третьего рода.


## Шаг 2: Пересчет уравнения и начальных условий

### Вычисление производных от $s(x,t)$

$$s(x,t) = x\sin t + \frac{x^2}{3}(2\cos t + e^t - 2\sin t)$$

**Вторая производная по $x$:**
$$s_{xx} = \frac{2}{3}(2\cos t + e^t - 2\sin t)$$

**Вторая производная по времени:**
$$s_{tt} = -x\sin t + \frac{x^2}{3}(-2\sin t + e^t - 2\cos t)$$

### Пересчет уравнения

Подставляем $u = v + s$ в уравнение $u_{tt} = u_{xx} + f(x,t)$:

$$(v + s)_{tt} = (v + s)_{xx} + f(x,t)$$
$$v_{tt} + s_{tt} = v_{xx} + s_{xx} + f(x,t)$$

Переносим все слагаемые с $s$ в правую часть:
$$v_{tt} = v_{xx} + f(x,t) - s_{tt} + s_{xx} = v_{xx} + \tilde{f}(x,t)$$

где:
$$\tilde{f}(x,t) = f(x,t) - s_{tt} + s_{xx} = [2\sin\frac{3\pi x}{2} + xe^{-t}] - s_{tt} + s_{xx}$$

### Пересчет граничных условий

**Для левой границы:**
$$v_x(0,t) = u_x(0,t) - s_x(0,t) = \sin t - \sin t = 0$$ ✓

**Для правой границы:**
$$v(1,t) + v_x(1,t) = [u(1,t) + u_x(1,t)] - [s(1,t) + s_x(1,t)] = (2\cos t + e^t) - (2\cos t + e^t) = 0$$ ✓


### Пересчет начальных условий

**Начальное смещение:**
$$\tilde{\theta}(x) = u(x,0) - s(x,0) = \sin\frac{3\pi x}{2} + x + 1 - \frac{x^2}{3}(2 + 1 - 0) = \sin\frac{3\pi x}{2} + x + 1 - x^2$$

**Начальная скорость:**
$$\tilde{\psi}(x) = u_t(x,0) - s_t(x,0) = -\sin\frac{5\pi x}{2} - x + 1 - \frac{x^2}{3}(-2 + 1) = -\sin\frac{5\pi x}{2} - x + 1 - \frac{x^2}{3}(-1) = -\sin\frac{5\pi x}{2} - x + 1 + \frac{x^2}{3}$$

### Итог шага 2

Задача для функции $v(x,t)$ с однородными граничными условиями третьего рода:

$$
\begin{cases}
v_{tt} = v_{xx} + \tilde{f}(x,t), & x \in [0,1], \quad t \geq 0 \\
v_x(0,t) = 0, \quad v(1,t) + v_x(1,t) = 0 \\
v(x,0) = \sin\frac{3\pi x}{2} + x + 1 - x^2, \quad v_t(x,0) = -\sin\frac{5\pi x}{2} - x + 1 + \frac{x^2}{3}
\end{cases}
$$

где $\tilde{f}(x,t)$ вычисляется через $f(x,t)$, $s_{tt}$ и $s_{xx}$.


## Шаг 3: Определение собственных функций и собственных значений

### Разделение переменных

Рассматриваем однородную задачу (без неоднородности):
$$v_{tt} = v_{xx}, \quad v_x(0,t) = 0, \quad v(1,t) + v_x(1,t) = 0$$

Ищем решение в виде $v(x,t) = X(x) \cdot T(t)$.

Подставляем в уравнение:
$$X(x) \cdot T''(t) = X''(x) \cdot T(t)$$

Разделяя переменные:
$$\frac{T''(t)}{T(t)} = \frac{X''(x)}{X(x)} = -\lambda^2$$

### Пространственное уравнение (задача Штурма-Лиувилля)

Получаем задачу Штурма-Лиувилля:
$$X''(x) + \lambda^2 X(x) = 0$$

**Граничные условия третьего рода:**
- $X'(0) = 0$
- $X(1) + X'(1) = 0$

### Общее решение пространственного уравнения

**Общее решение:**
$$X(x) = c_1 \sin\lambda x + c_2 \cos\lambda x$$

**Производная:**
$$X'(x) = c_1 \lambda \cos\lambda x - c_2 \lambda \sin\lambda x$$


### Применение граничных условий

**Первое условие:** $X'(0) = 0$

Подставляем $x = 0$:
$$X'(0) = c_1 \lambda \cos(0) - c_2 \lambda \sin(0) = c_1 \lambda \cdot 1 - 0 = c_1 \lambda = 0$$

Для нетривиального решения ($\lambda \neq 0$): $c_1 = 0$

Следовательно: $X(x) = c_2 \cos\lambda x$

**Второе условие:** $X(1) + X'(1) = 0$

Подставляем $x = 1$:
$$X(1) + X'(1) = c_2 \cos\lambda + (-c_2 \lambda \sin\lambda) = c_2(\cos\lambda - \lambda \sin\lambda) = 0$$

Для нетривиального решения ($c_2 \neq 0$):
$$\cos\lambda - \lambda \sin\lambda = 0$$

### Трансцендентное уравнение для собственных значений

Разделим обе части на $\cos\lambda$ (при условии, что $\cos\lambda \neq 0$):
$$1 - \lambda \tan\lambda = 0$$

Откуда получаем трансцендентное уравнение:
$$\tan\lambda = \frac{1}{\lambda}$$

**Важно:** Это уравнение не решается аналитически. Его корни $\lambda_k$ ($k = 1, 2, 3, \ldots$) находятся численно.

### Собственные функции и собственные значения

**Собственные функции:**
$$X_k(x) = \cos(\lambda_k x), \quad k = 1, 2, 3, \ldots$$

**Собственные значения:** $\lambda_k$ — положительные корни уравнения $\tan\lambda = \frac{1}{\lambda}$

**Первые корни (приближенно):**
- $\lambda_1 \approx 0.8603$
- $\lambda_2 \approx 3.4256$
- $\lambda_3 \approx 6.4373$
- $\lambda_4 \approx 9.5293$
- и т.д.

### Итог шага 3

Мы нашли:
- **Собственные значения:** $\lambda_k$ — положительные корни уравнения $\tan\lambda = \frac{1}{\lambda}$ (вычисляются численно)
- **Собственные функции:** $X_k(x) = \cos(\lambda_k x)$, $k = 1, 2, 3, \ldots$

Каждая функция $X_k(x)$ удовлетворяет:
- Уравнению $X_k''(x) + \lambda_k^2 X_k(x) = 0$
- Граничным условиям $X_k'(0) = 0$ и $X_k(1) + X_k'(1) = 0$


## Шаг 4: Разложение функций в ряды Фурье

### Представление решения в виде ряда

Ищем решение в виде бесконечного ряда по собственным функциям:
$$v(x,t) = \sum_{k=1}^{\infty} X_k(x) T_k(t) = \sum_{k=1}^{\infty} \cos(\lambda_k x) T_k(t)$$

### Норма собственных функций

Для вычисления коэффициентов Фурье нужно вычислить норму собственных функций:
$$\|X_k\|^2 = \int_0^1 \cos^2(\lambda_k x) dx = \frac{1}{2}\int_0^1 (1 + \cos(2\lambda_k x)) dx$$

$$= \frac{1}{2}\left[x + \frac{\sin(2\lambda_k x)}{2\lambda_k}\right]_0^1 = \frac{1}{2} + \frac{\sin(2\lambda_k)}{4\lambda_k}$$

### Разложение начального смещения

Разложим функцию $\tilde{\theta}(x) = \sin\frac{3\pi x}{2} + x + 1 - x^2$ в ряд Фурье:

$$\theta_k = \frac{1}{\|X_k\|^2}\int_0^1 \left[\sin\frac{3\pi x}{2} + x + 1 - x^2\right] \cos(\lambda_k x) dx$$

### Разложение начальной скорости

Разложим функцию $\tilde{\psi}(x) = -\sin\frac{5\pi x}{2} - x + 1 + \frac{x^2}{3}$ в ряд Фурье:

$$\psi_k = \frac{1}{\|X_k\|^2}\int_0^1 \left[-\sin\frac{5\pi x}{2} - x + 1 + \frac{x^2}{3}\right] \cos(\lambda_k x) dx$$


### Разложение неоднородности

Разложим неоднородность $\tilde{f}(x,t)$:

$$\mu_k(t) = \frac{1}{\|X_k\|^2}\int_0^1 \tilde{f}(x,t) \cos(\lambda_k x) dx$$

где $\tilde{f}(x,t) = f(x,t) - s_{tt} + s_{xx}$ вычисляется на шаге 2.

### Итог шага 4

**Разложение начального смещения:**
$$\sin\frac{3\pi x}{2} + x + 1 - x^2 = \sum_{k=1}^{\infty} \theta_k \cos(\lambda_k x)$$

**Разложение начальной скорости:**
$$-\sin\frac{5\pi x}{2} - x + 1 + \frac{x^2}{3} = \sum_{k=1}^{\infty} \psi_k \cos(\lambda_k x)$$

**Разложение неоднородности:**
$$\tilde{f}(x,t) = \sum_{k=1}^{\infty} \mu_k(t) \cos(\lambda_k x)$$

где все коэффициенты вычисляются через интегралы с весами $\frac{1}{\|X_k\|^2}$.


## Шаг 5: Решение уравнений для $T_k(t)$

### Подстановка ряда в уравнение

Подставляем ряд $v(x,t) = \sum_{k=1}^{\infty} \cos(\lambda_k x) T_k(t)$ в уравнение $v_{tt} = v_{xx} + \tilde{f}(x,t)$:

**Левая часть:**
$$\frac{\partial^2}{\partial t^2}\left[\sum_{k=1}^{\infty} \cos(\lambda_k x) T_k(t)\right] = \sum_{k=1}^{\infty} \cos(\lambda_k x) T_k''(t)$$

**Правая часть:**
$$\frac{\partial^2}{\partial x^2}\left[\sum_{k=1}^{\infty} \cos(\lambda_k x) T_k(t)\right] + \tilde{f}(x,t)$$

Используя $X_k''(x) = -\lambda_k^2 X_k(x) = -\lambda_k^2 \cos(\lambda_k x)$:
$$= \sum_{k=1}^{\infty} (-\lambda_k^2) \cos(\lambda_k x) T_k(t) + \sum_{k=1}^{\infty} \mu_k(t) \cos(\lambda_k x)$$

### Приравнивание коэффициентов

Приравниваем коэффициенты при одинаковых функциях $\cos(\lambda_k x)$:

$$T_k''(t) = -\lambda_k^2 T_k(t) + \mu_k(t), \quad k = 1, 2, \ldots$$

**Начальные условия:** $T_k(0) = \theta_k$, $T_k'(0) = \psi_k$


### Решение неоднородного уравнения методом вариации постоянных

Это линейное неоднородное ОДУ второго порядка:
$$T_k''(t) + \lambda_k^2 T_k(t) = \mu_k(t)$$

**Решение методом вариации постоянных:**
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

### Итог шага 5

Решение временных уравнений:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

где $\lambda_k$ — собственные значения (корни уравнения $\tan\lambda = \frac{1}{\lambda}$), а $\theta_k$, $\psi_k$ и $\mu_k(t)$ найдены на шаге 4.


## Шаг 6: Запись полного решения для $v(x,t)$

### Общая формула

На шаге 4 мы записали решение в виде ряда:
$$v(x,t) = \sum_{k=1}^{\infty} \cos(\lambda_k x) T_k(t)$$

Подставляем найденные выражения для $T_k(t)$ из шага 5:
$$v(x,t) = \sum_{k=1}^{\infty} \cos(\lambda_k x) \left[\theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau\right]$$

### Итог шага 6

Полное решение для функции $v(x,t)$:
$$v(x,t) = \sum_{k=1}^{\infty} \cos(\lambda_k x) T_k(t)$$

где $T_k(t)$ находятся из уравнений на шаге 5 и содержат осцилляции (колебания) с частотами $\lambda_k$.


## Шаг 7: Окончательный ответ

### Возврат к исходной функции

На шаге 1 мы установили: $u(x,t) = v(x,t) + s(x,t)$

Подставляем:
- $v(x,t)$ из шага 6
- $s(x,t) = x\sin t + \frac{x^2}{3}(2\cos t + e^t - 2\sin t)$ из шага 1

### Финальный ответ

**Ответ задачи №7:**

$$u(x,t) = v(x,t) + s(x,t) = \sum_{k=1}^{\infty} T_k(t)\cos(\lambda_k x) + x\sin t + \frac{x^2}{3}(2\cos t + e^t - 2\sin t)$$

где:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

$$\lambda_k$$ — положительные корни уравнения $\tan\lambda = \frac{1}{\lambda}$ (вычисляются численно)

а коэффициенты $\theta_k$, $\psi_k$ и $\mu_k(t)$ вычисляются через интегралы Фурье на шаге 4.

### Физическая интерпретация решения

Решение состоит из трех компонент:
1. **Свободные колебания струны** с собственными частотами $\lambda_k$ (гармоники), где $\lambda_k$ — корни трансцендентного уравнения
2. **Вынужденные колебания** от источника $\tilde{f}(x,t)$ (интегральные члены)
3. **Вклад от неоднородных граничных условий третьего рода:** $x\sin t + \frac{x^2}{3}(2\cos t + e^t - 2\sin t)$

**Особенность:** Собственные значения $\lambda_k$ не выражаются через элементарные функции (трансцендентное уравнение), поэтому решение требует численных методов.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import fsolve
from scipy.integrate import quad, odeint

# Находим собственные значения (корни уравнения tan(λ) = 1/λ)
def transcendental_eq(lam):
    if abs(lam) < 1e-10:
        return 1
    return np.tan(lam) - 1/lam

# Найдем первые N собственных значений
N_TERMS = 10
eigenvalues = []
for k in range(1, N_TERMS + 1):
    # Начальное приближение для k-го корня
    initial_guess = (k - 0.5) * np.pi + 0.1
    try:
        lam_k = fsolve(transcendental_eq, initial_guess)[0]
        if lam_k > 0:  # Берем только положительные корни
            eigenvalues.append(lam_k)
    except:
        pass

print(f"Первые {len(eigenvalues)} собственных значений:")
for i, lam in enumerate(eigenvalues):
    print(f"λ_{i+1} = {lam:.6f}")

# Параметры визуализации
x = np.linspace(0, 1, 100)
t_surf = np.linspace(0, 2, 50)

# Функция s(x,t)
def s_7(x, t):
    return x * np.sin(t) + (x**2 / 3) * (2*np.cos(t) + np.exp(t) - 2*np.sin(t))

# Норма собственных функций
def norm_squared(lam):
    return 0.5 + np.sin(2*lam)/(4*lam)

# Начальные условия после приведения к однородным граничным условиям
def theta_tilde(x):
    return np.sin(3*np.pi*x/2) + x + 1 - x**2

def psi_tilde(x):
    return -np.sin(5*np.pi*x/2) - x + 1 + x**2/3

# Вычисление коэффициентов Фурье
theta_k = []
psi_k = []

for lam in eigenvalues:
    norm2 = norm_squared(lam)
    # Коэффициенты разложения начальных условий
    theta = quad(lambda x: theta_tilde(x) * np.cos(lam*x), 0, 1)[0] / norm2
    psi = quad(lambda x: psi_tilde(x) * np.cos(lam*x), 0, 1)[0] / norm2
    theta_k.append(theta)
    psi_k.append(psi)

# Упрощенное решение для визуализации (без учета неоднородного члена для упрощения)
def u_7_analytical(x_val, t_val):
    v = sum(theta_k[i] * np.cos(eigenvalues[i] * t_val) * np.cos(eigenvalues[i] * x_val) 
            + psi_k[i]/eigenvalues[i] * np.sin(eigenvalues[i] * t_val) * np.cos(eigenvalues[i] * x_val)
            for i in range(len(eigenvalues)))
    return v + s_7(x_val, t_val)

# Визуализация
fig = plt.figure(figsize=(16, 7))

# График 1: u(x,t) при различных t
ax1 = fig.add_subplot(121)
t_values = [0, 0.5, 1.0, 1.5, 2.0]
for t in t_values:
    u = [u_7_analytical(xi, t) for xi in x]
    ax1.plot(x, u, label=f'$t = {t:.1f}$', linewidth=2)

ax1.set_xlabel('$x$', fontsize=12)
ax1.set_ylabel('$u(x,t)$', fontsize=12)
ax1.set_title(f'Задача №7: $u(x,t)$ при различных $t$ (N={len(eigenvalues)})', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim([0, 1])

# График 2: 3D поверхность u(x,t)
ax2 = fig.add_subplot(122, projection='3d')
X, T = np.meshgrid(x, t_surf)
U = np.array([[u_7_analytical(xi, ti) for xi in x] for ti in t_surf])

surf = ax2.plot_surface(X, T, U, cmap='viridis', alpha=0.9)
ax2.set_xlabel('$x$'); ax2.set_ylabel('$t$'); ax2.set_zlabel('$u(x,t)$')
ax2.set_title(f'Задача №7: Поверхность $u(x,t)$ (граничные условия III рода)')
fig.colorbar(surf, ax=ax2, shrink=0.5)
plt.tight_layout(); plt.show()
print("Визуализация задачи №7 с граничными условиями третьего рода")
```
