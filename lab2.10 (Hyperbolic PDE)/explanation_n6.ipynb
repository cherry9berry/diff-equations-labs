# Подробный разбор решения Задачи №6 (Гиперболическое уравнение со смешанными граничными условиями)

## Постановка задачи

**Дано:**
- **Уравнение:** $u_{tt} = u_{xx}$ (однородное уравнение, $f(x,t) = 0$), $x \in [0,1]$, $t \geq 0$
- **Граничные условия:**
  - $u(0,t) = e^t$ (левая граница экспоненциально растет)
  - $u_x(1,t) = 3e^t$ (производная на правой границе экспоненциально растет)
- **Начальные условия:**
  - $u(x,0) = 2\sin\frac{3\pi x}{2}$ (начальное смещение)
  - $u_t(x,0) = -3\sin\frac{5\pi x}{2}$ (начальная скорость)

**Особенность:** Смешанные граничные условия (условие Дирихле на левом конце, условие Неймана на правом конце). Исходное уравнение однородное, но неоднородность возникает из-за несоответствия граничных условий.


## Шаг 1: Приведение граничных условий к однородным

### Зачем это нужно?

Граничные условия **неоднородные** (правые части $e^t$ и $3e^t$ не равны нулю), поэтому нужно найти функцию $s(x,t)$, которая "возьмет на себя" эти неоднородности.

### Поиск функции $s(x,t)$

Представим $u(x,t) = v(x,t) + s(x,t)$, где $s(x,t)$ должна удовлетворять:
- $s(0,t) = e^t$
- $s_x(1,t) = 3e^t$

### Подбор функции $s(x,t)$

Попробуем найти $s(x,t)$ в виде линейной функции по $x$:
$$s(x,t) = (\alpha_0 + \alpha_1 x)A(t)$$

где $\alpha_0$, $\alpha_1$ — константы, $A(t)$ — функция времени.

**Вычисляем производную:**
$$s_x(x,t) = \alpha_1 A(t)$$

**Применяем граничные условия:**

Из первого условия: $s(0,t) = \alpha_0 A(t) = e^t$

Положим $\alpha_0 = 1$, $A(t) = e^t$.

Из второго условия: $s_x(1,t) = \alpha_1 e^t = 3e^t \Rightarrow \alpha_1 = 3$

### Результат

$$s(x,t) = (1 + 3x)e^t$$

### Проверка граничных условий

- $s(0,t) = (1 + 0)e^t = e^t$ ✓
- $s_x(x,t) = 3e^t$, следовательно $s_x(1,t) = 3e^t$ ✓

**Итог шага 1:** Функция $s(x,t) = (1 + 3x)e^t$ удовлетворяет граничным условиям.


## Шаг 2: Пересчет уравнения и начальных условий

### Вычисление производных от $s(x,t) = (1 + 3x)e^t$

**Первая производная по $x$:**
$$s_x = 3e^t$$

**Вторая производная по $x$:**
$$s_{xx} = 0$$

**Первая производная по времени:**
$$s_t = (1 + 3x)e^t$$

**Вторая производная по времени:**
$$s_{tt} = (1 + 3x)e^t$$

### Пересчет уравнения

Подставляем $u = v + s$ в уравнение $u_{tt} = u_{xx}$:

$$(v + s)_{tt} = (v + s)_{xx}$$
$$v_{tt} + s_{tt} = v_{xx} + s_{xx}$$

Переносим все слагаемые с $s$ в правую часть:
$$v_{tt} = v_{xx} + s_{xx} - s_{tt} = v_{xx} + 0 - (1 + 3x)e^t$$

**Результат:** Уравнение для $v(x,t)$:
$$v_{tt} = v_{xx} - (1 + 3x)e^t$$

где $\tilde{f}(x,t) = -(1 + 3x)e^t$ — неоднородность, возникающая из-за несоответствия $s_{tt}$ и $s_{xx}$ (хотя исходное уравнение было однородным).


### Пересчет граничных условий

**Для левой границы:**
$$v(0,t) = u(0,t) - s(0,t) = e^t - e^t = 0$$ ✓

**Для правой границы:**
$$v_x(1,t) = u_x(1,t) - s_x(1,t) = 3e^t - 3e^t = 0$$ ✓

### Пересчет начальных условий

**Начальное смещение:**
$$\tilde{\theta}(x) = u(x,0) - s(x,0) = 2\sin\frac{3\pi x}{2} - (1 + 3x) \cdot 1 = 2\sin\frac{3\pi x}{2} - 1 - 3x$$

**Начальная скорость:**
$$\tilde{\psi}(x) = u_t(x,0) - s_t(x,0) = -3\sin\frac{5\pi x}{2} - (1 + 3x) \cdot 1 = -3\sin\frac{5\pi x}{2} - 1 - 3x$$

### Итог шага 2

Задача для функции $v(x,t)$ с однородными граничными условиями:

$$
\begin{cases}
v_{tt} = v_{xx} - (1 + 3x)e^t, & x \in [0,1], \quad t \geq 0 \\
v(0,t) = 0, \quad v_x(1,t) = 0 \\
v(x,0) = 2\sin\frac{3\pi x}{2} - 1 - 3x, \quad v_t(x,0) = -3\sin\frac{5\pi x}{2} - 1 - 3x
\end{cases}
$$

где:
- $\tilde{f}(x,t) = -(1 + 3x)e^t$ — неоднородность
- $\tilde{\theta}(x) = 2\sin\frac{3\pi x}{2} - 1 - 3x$ — начальное смещение
- $\tilde{\psi}(x) = -3\sin\frac{5\pi x}{2} - 1 - 3x$ — начальная скорость


## Шаг 3: Определение собственных функций и собственных значений

### Разделение переменных

Рассматриваем однородную задачу (без неоднородности):
$$v_{tt} = v_{xx}, \quad v(0,t) = 0, \quad v_x(1,t) = 0$$

Ищем решение в виде $v(x,t) = X(x) \cdot T(t)$.

Подставляем в уравнение:
$$X(x) \cdot T''(t) = X''(x) \cdot T(t)$$

Разделяя переменные:
$$\frac{T''(t)}{T(t)} = \frac{X''(x)}{X(x)} = -\lambda^2$$

### Пространственное уравнение

Получаем задачу Штурма-Лиувилля:
$$X''(x) + \lambda^2 X(x) = 0$$

**Граничные условия:** $X(0) = 0$ и $X'(1) = 0$ (смешанные условия)

### Общее решение пространственного уравнения

**Общее решение:**
$$X(x) = c_1 \sin\lambda x + c_2 \cos\lambda x$$

**Производная:**
$$X'(x) = c_1 \lambda \cos\lambda x - c_2 \lambda \sin\lambda x$$


### Применение граничных условий

**Первое условие:** $X(0) = 0$

Подставляем $x = 0$:
$$X(0) = c_1 \sin(0) + c_2 \cos(0) = 0 + c_2 \cdot 1 = c_2 = 0$$

Следовательно: $X(x) = c_1 \sin\lambda x$

**Второе условие:** $X'(1) = 0$

Подставляем $x = 1$:
$$X'(1) = c_1 \lambda \cos\lambda = 0$$

Для нетривиального решения ($c_1 \neq 0$): $\cos\lambda = 0$

### Нахождение собственных значений

Уравнение $\cos\lambda = 0$ имеет решения:
$$\lambda = \frac{\pi}{2} + n\pi = \frac{\pi(2n+1)}{2}, \quad n = 0, 1, 2, \ldots$$

Обозначая $k = n + 1 = 1, 2, 3, \ldots$:

**Собственные значения:**
$$\lambda_k = \frac{\pi(2k-1)}{2}, \quad k = 1, 2, 3, \ldots$$

### Собственные функции

Для каждого собственного значения $\lambda_k$ получаем соответствующую собственную функцию:
$$X_k(x) = \sin\left(\frac{\pi(2k-1)x}{2}\right), \quad k = 1, 2, 3, \ldots$$

**Проверка:** При $k=2$ получаем $\lambda_2 = \frac{3\pi}{2}$, что соответствует начальному условию $2\sin\frac{3\pi x}{2}$.

### Итог шага 3

Мы нашли:
- **Собственные значения:** $\lambda_k = \frac{\pi(2k-1)}{2}$, $k = 1, 2, 3, \ldots$
- **Собственные функции:** $X_k(x) = \sin\left(\frac{\pi(2k-1)x}{2}\right)$, $k = 1, 2, 3, \ldots$

Каждая функция $X_k(x)$ удовлетворяет:
- Уравнению $X_k''(x) + \lambda_k^2 X_k(x) = 0$
- Граничным условиям $X_k(0) = 0$ и $X_k'(1) = 0$


## Шаг 4: Разложение функций в ряды Фурье

### Представление решения в виде ряда

Ищем решение в виде бесконечного ряда по собственным функциям:
$$v(x,t) = \sum_{k=1}^{\infty} X_k(x) T_k(t) = \sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t)$$

### Разложение начального смещения

Разложим функцию $\tilde{\theta}(x) = 2\sin\frac{3\pi x}{2} - 1 - 3x$ в ряд Фурье по системе собственных функций $\left\{\sin\left(\frac{\pi(2k-1)x}{2}\right)\right\}$.

**Коэффициенты Фурье** вычисляются по формуле:
$$\theta_k = 2\int_0^1 \left[2\sin\frac{3\pi x}{2} - 1 - 3x\right] \sin\left(\frac{\pi(2k-1)x}{2}\right)dx$$

**Важное наблюдение:** При $k=2$ (т.е. $\lambda_2 = \frac{3\pi}{2}$): первый член даст $\theta_2 = 2$, остальные члены (от $-1 - 3x$) вычисляются численно.

### Разложение начальной скорости

Разложим функцию $\tilde{\psi}(x) = -3\sin\frac{5\pi x}{2} - 1 - 3x$ в ряд Фурье.

**Коэффициенты Фурье:**
$$\psi_k = 2\int_0^1 \left[-3\sin\frac{5\pi x}{2} - 1 - 3x\right] \sin\left(\frac{\pi(2k-1)x}{2}\right)dx$$

**Важное наблюдение:** При $k=3$ (т.е. $\lambda_3 = \frac{5\pi}{2}$): первый член даст $\psi_3 = -3$, остальные вычисляются численно.


### Разложение неоднородности

Теперь разложим неоднородность $\tilde{f}(x,t) = -(1 + 3x)e^t$:

$$\tilde{f}(x,t) = \sum_{k=1}^{\infty} \mu_k(t) X_k(x) = \sum_{k=1}^{\infty} \mu_k(t) \sin\left(\frac{\pi(2k-1)x}{2}\right)$$

**Коэффициенты Фурье** вычисляются по формуле:
$$\mu_k(t) = 2\int_0^1 [-(1 + 3x)e^t] \sin\left(\frac{\pi(2k-1)x}{2}\right)dx$$
$$= -2e^t \int_0^1 (1 + 3x) \sin\left(\frac{\pi(2k-1)x}{2}\right)dx$$

Обозначим:
$$I_k = \int_0^1 (1 + 3x) \sin\left(\frac{\pi(2k-1)x}{2}\right)dx$$

Тогда: $\mu_k(t) = -2e^t I_k$

**Примечание:** Интегралы $I_k$ вычисляются численно.

### Итог шага 4

**Разложение начального смещения:**
$$2\sin\frac{3\pi x}{2} - 1 - 3x = \sum_{k=1}^{\infty} \theta_k \sin\left(\frac{\pi(2k-1)x}{2}\right)$$

где $\theta_2 = 2$ (главный член), остальные вычисляются численно.

**Разложение начальной скорости:**
$$-3\sin\frac{5\pi x}{2} - 1 - 3x = \sum_{k=1}^{\infty} \psi_k \sin\left(\frac{\pi(2k-1)x}{2}\right)$$

где $\psi_3 = -3$ (главный член), остальные вычисляются численно.

**Разложение неоднородности:**
$$\tilde{f}(x,t) = \sum_{k=1}^{\infty} \mu_k(t) \sin\left(\frac{\pi(2k-1)x}{2}\right)$$

где $\mu_k(t) = -2e^t I_k$ и зависят от времени через экспоненту.


## Шаг 5: Решение уравнений для $T_k(t)$

### Подстановка ряда в уравнение

Подставляем ряд $v(x,t) = \sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t)$ в уравнение $v_{tt} = v_{xx} - (1 + 3x)e^t$:

**Левая часть:**
$$\frac{\partial^2}{\partial t^2}\left[\sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t)\right] = \sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k''(t)$$

**Правая часть:**
$$\frac{\partial^2}{\partial x^2}\left[\sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t)\right] + \tilde{f}(x,t)$$

Используя $X_k''(x) = -\lambda_k^2 X_k(x)$:
$$= \sum_{k=1}^{\infty} (-\lambda_k^2) \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t) + \sum_{k=1}^{\infty} \mu_k(t) \sin\left(\frac{\pi(2k-1)x}{2}\right)$$

### Приравнивание коэффициентов

Приравниваем коэффициенты при одинаковых функциях $\sin\left(\frac{\pi(2k-1)x}{2}\right)$:

$$T_k''(t) = -\lambda_k^2 T_k(t) + \mu_k(t), \quad k = 1, 2, \ldots$$

**Начальные условия:** $T_k(0) = \theta_k$, $T_k'(0) = \psi_k$


### Решение неоднородного уравнения методом вариации постоянных

Это линейное неоднородное ОДУ второго порядка:
$$T_k''(t) + \lambda_k^2 T_k(t) = \mu_k(t)$$

**Решение методом вариации постоянных:**
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

Подставляя $\mu_k(t) = -2e^t I_k$:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) - \frac{2I_k}{\lambda_k}\int_0^t e^\tau\sin[\lambda_k(t-\tau)] d\tau$$

**Примечание:** Интеграл $\int_0^t e^\tau\sin[\lambda_k(t-\tau)] d\tau$ вычисляется численно или аналитически, но в общем случае используется численный метод.

### Итог шага 5

Решение временных уравнений:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

где $\lambda_k = \frac{\pi(2k-1)}{2}$, а $\theta_k$, $\psi_k$ и $\mu_k(t) = -2e^t I_k$ найдены на шаге 4.


## Шаг 6: Запись полного решения для $v(x,t)$

### Общая формула

На шаге 4 мы записали решение в виде ряда:
$$v(x,t) = \sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t)$$

Подставляем найденные выражения для $T_k(t)$ из шага 5:
$$v(x,t) = \sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) \left[\theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau\right]$$

### Итог шага 6

Полное решение для функции $v(x,t)$:
$$v(x,t) = \sum_{k=1}^{\infty} \sin\left(\frac{\pi(2k-1)x}{2}\right) T_k(t)$$

где $T_k(t)$ находятся из уравнений на шаге 5 и содержат осцилляции (колебания) с частотами $\lambda_k = \frac{\pi(2k-1)}{2}$.


## Шаг 7: Окончательный ответ

### Возврат к исходной функции

На шаге 1 мы установили: $u(x,t) = v(x,t) + s(x,t)$

Подставляем:
- $v(x,t)$ из шага 6
- $s(x,t) = (1 + 3x)e^t$ из шага 1

### Финальный ответ

**Ответ задачи №6:**

$$u(x,t) = v(x,t) + s(x,t) = \sum_{k=1}^{\infty} T_k(t)\sin\left(\frac{\pi(2k-1)x}{2}\right) + (1 + 3x)e^t$$

где:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

$$\lambda_k = \frac{\pi(2k-1)}{2}$$

а коэффициенты $\theta_k$, $\psi_k$ и $\mu_k(t) = -2e^t I_k$ вычисляются через интегралы Фурье на шаге 4.

### Физическая интерпретация решения

Решение состоит из трех компонент:
1. **Свободные колебания струны** с собственными частотами $\lambda_k = \frac{\pi(2k-1)}{2}$ (гармоники)
2. **Вынужденные колебания** от источника $\tilde{f}(x,t) = -(1 + 3x)e^t$ (интегральные члены)
3. **Вклад от неоднородных граничных условий:** $(1 + 3x)e^t$ — экспоненциально растет со временем

**Важно:** Границы экспоненциально растут, что приводит к экспоненциальному росту решения.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad, odeint

# Параметры
x = np.linspace(0, 1, 100)
t_values = [0, 0.5, 1.0, 1.5, 2.0]
t_surf = np.linspace(0, 2, 50)
N_TERMS = 20

# Функция s(x,t)
def s_6(x, t):
    return (1 + 3*x) * np.exp(t)

# Собственные значения
def lambda_k(k):
    return np.pi * (2*k - 1) / 2

# Вычисление коэффициентов Фурье
theta_k = np.zeros(N_TERMS + 1)
psi_k = np.zeros(N_TERMS + 1)
I_k = np.zeros(N_TERMS + 1)

for k in range(1, N_TERMS + 1):
    lam = lambda_k(k)
    
    # theta_k: разложение [2*sin(3*pi*x/2) - 1 - 3*x]
    integrand_theta = lambda x: (2*np.sin(3*np.pi*x/2) - 1 - 3*x) * np.sin(lam * x)
    theta_k[k] = 2 * quad(integrand_theta, 0, 1)[0]
    
    # psi_k: разложение [-3*sin(5*pi*x/2) - 1 - 3*x]
    integrand_psi = lambda x: (-3*np.sin(5*np.pi*x/2) - 1 - 3*x) * np.sin(lam * x)
    psi_k[k] = 2 * quad(integrand_psi, 0, 1)[0]
    
    # I_k для mu_k(t) = -2*e^t * I_k
    integrand_I = lambda x: (1 + 3*x) * np.sin(lam * x)
    I_k[k] = quad(integrand_I, 0, 1)[0]

# Функция mu_k(t)
def mu_k_6(t, k):
    return -2 * np.exp(t) * I_k[k]

# Модель ОДУ для T_k(t)
def model_6(Y, t, k):
    T, T_prime = Y
    lam = lambda_k(k)
    dYdt = [T_prime, mu_k_6(t, k) - lam**2 * T]
    return dYdt

# Решение ОДУ для всех k
T_solutions_6 = []
for k in range(1, N_TERMS + 1):
    Y0 = [theta_k[k], psi_k[k]]
    sol = odeint(model_6, Y0, t_surf, args=(k,))
    T_solutions_6.append(sol[:, 0])

# Вычисление u(x,t)
def u_6_analytical(x, t, T_snapshot):
    s_xt = s_6(x, t)
    v_xt = sum(T_k * np.sin(lambda_k(k+1) * x) for k, T_k in enumerate(T_snapshot))
    return v_xt + s_xt

# Создание фигуры
fig = plt.figure(figsize=(16, 6))

# График 1: Зависимость от x при разных t
ax1 = fig.add_subplot(121)
for t_val in t_values:
    t_idx = np.abs(t_surf - t_val).argmin()
    T_snapshot = [T_sol[t_idx] for T_sol in T_solutions_6]
    u = u_6_analytical(x, t_surf[t_idx], T_snapshot)
    ax1.plot(x, u, label=f'$t = {t_val:.1f}$')
ax1.set_xlabel('$x$', fontsize=12)
ax1.set_ylabel('$u(x,t)$', fontsize=12)
ax1.set_title(f'Задача №6: $u(x,t)$ при различных $t$ (N={N_TERMS})', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim([0, 1])

# График 2: 3D поверхность u(x,t)
ax2 = fig.add_subplot(122, projection='3d')
X, T = np.meshgrid(x, t_surf)
U = np.array([[u_6_analytical(xi, t_surf[i], [T_sol[i] for T_sol in T_solutions_6]) for xi in x] for i in range(len(t_surf))])

surf = ax2.plot_surface(X, T, U, cmap='viridis', alpha=0.9, linewidth=0, antialiased=True)
ax2.set_xlabel('$x$', fontsize=12)
ax2.set_ylabel('$t$', fontsize=12)
ax2.set_zlabel('$u(x,t)$', fontsize=12)
ax2.set_title(f'Задача №6: Поверхность $u(x,t)$ (N={N_TERMS})', fontsize=14)
fig.colorbar(surf, ax=ax2, shrink=0.5, aspect=5, label='$u(x,t)$')

plt.tight_layout()
plt.show()
print("Визуализация задачи №6")
```
