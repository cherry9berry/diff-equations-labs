# Подробный разбор решения Задачи №5 (Гиперболическое уравнение)

## Постановка задачи

**Дано:**
- **Уравнение:** $u_{tt} = u_{xx} + f(x,t)$, где $f(x,t) = -\sin\pi x - x(\sin t - \cos t) - \sin t$, $x \in [0,1]$, $t \geq 0$
- **Граничные условия:**
  - $u(0,t) = \cos t$ (левая граница колеблется по косинусу)
  - $u(1,t) = \sin t$ (правая граница колеблется по синусу)
- **Начальные условия:**
  - $u(x,0) = 3\sin\pi x - x + 1$ (начальное смещение)
  - $u_t(x,0) = -\sin 2\pi x + x$ (начальная скорость)

**Особенность:** Это гиперболическое уравнение (волновое уравнение), которое описывает колебания струны или другую волновую динамику. В отличие от параболических уравнений, здесь есть две производные по времени, и решение содержит осцилляции (колебания), а не только затухание.


## Шаг 1: Приведение граничных условий к однородным

### Зачем это нужно?

Граничные условия **неоднородные** (правые части $\cos t$ и $\sin t$ не равны нулю), поэтому нужно найти функцию $s(x,t)$, которая "возьмет на себя" эти неоднородности.

### Поиск функции $s(x,t)$

Представим $u(x,t) = v(x,t) + s(x,t)$, где $s(x,t)$ должна удовлетворять:
- $s(0,t) = \cos t$
- $s(1,t) = \sin t$

### Линейная интерполяция

Используем линейную интерполяцию между граничными значениями:
$$s(x,t) = (1-x)g_1(t) + xg_2(t)$$

где $g_1(t) = \cos t$ (значение на левой границе) и $g_2(t) = \sin t$ (значение на правой границе).

### Результат

$$s(x,t) = (1-x)\cos t + x\sin t$$

### Проверка граничных условий

**Левая граница** ($x = 0$):
$$s(0,t) = (1-0)\cos t + 0 \cdot \sin t = \cos t$$ ✓

**Правая граница** ($x = 1$):
$$s(1,t) = (1-1)\cos t + 1 \cdot \sin t = \sin t$$ ✓

**Итог шага 1:** Функция $s(x,t) = (1-x)\cos t + x\sin t$ удовлетворяет граничным условиям.


## Шаг 2: Пересчет уравнения и начальных условий

### Вычисление производных от $s(x,t) = (1-x)\cos t + x\sin t$

**Первая производная по $x$:**
$$s_x = -\cos t + \sin t$$

**Вторая производная по $x$:**
$$s_{xx} = 0$$

**Первая производная по времени:**
$$s_t = -(1-x)\sin t + x\cos t$$

**Вторая производная по времени:**
$$s_{tt} = -(1-x)\cos t - x\sin t$$

### Пересчет уравнения

Подставляем $u = v + s$ в уравнение $u_{tt} = u_{xx} + f(x,t)$:

$$(v + s)_{tt} = (v + s)_{xx} + f(x,t)$$
$$v_{tt} + s_{tt} = v_{xx} + s_{xx} + f(x,t)$$

Переносим все слагаемые с $s$ в правую часть:
$$v_{tt} = v_{xx} + f(x,t) + s_{xx} - s_{tt}$$

Подставляем найденные производные:
$$v_{tt} = v_{xx} + f(x,t) + 0 - [-(1-x)\cos t - x\sin t]$$
$$v_{tt} = v_{xx} + f(x,t) + (1-x)\cos t + x\sin t$$


### Упрощение неоднородности

Подставляем $f(x,t) = -\sin\pi x - x(\sin t - \cos t) - \sin t$:

$$\tilde{f}(x,t) = f(x,t) - s_{tt} = [-\sin\pi x - x(\sin t - \cos t) - \sin t] - [-(1-x)\cos t - x\sin t]$$

Раскрываем скобки:
$$= -\sin\pi x - x\sin t + x\cos t - \sin t + (1-x)\cos t + x\sin t$$

Упрощаем:
$$= -\sin\pi x - x\sin t + x\cos t - \sin t + \cos t - x\cos t + x\sin t$$
$$= -\sin\pi x - \sin t + \cos t$$

**Результат:** Уравнение для $v(x,t)$:
$$v_{tt} = v_{xx} + \tilde{f}(x,t)$$

где $\tilde{f}(x,t) = -\sin\pi x + \cos t - \sin t$.

### Пересчет граничных условий

**Левая граница:**
$$v(0,t) = u(0,t) - s(0,t) = \cos t - \cos t = 0$$ ✓

**Правая граница:**
$$v(1,t) = u(1,t) - s(1,t) = \sin t - \sin t = 0$$ ✓

### Пересчет начальных условий

**Начальное смещение:**
$$\tilde{\theta}(x) = u(x,0) - s(x,0) = [3\sin\pi x - x + 1] - [(1-x) \cdot 1 + x \cdot 0]$$
$$= 3\sin\pi x - x + 1 - 1 + x = 3\sin\pi x$$

**Начальная скорость:**
$$\tilde{\psi}(x) = u_t(x,0) - s_t(x,0) = [-\sin 2\pi x + x] - [-(1-x) \cdot 0 + x \cdot 1]$$
$$= -\sin 2\pi x + x - x = -\sin 2\pi x$$


### Итог шага 2

Задача для функции $v(x,t)$ с однородными граничными условиями:

$$
\begin{cases}
v_{tt} = v_{xx} + \tilde{f}(x,t), & x \in [0,1], \quad t \geq 0 \\
v(0,t) = 0, \quad v(1,t) = 0 \\
v(x,0) = 3\sin\pi x, \quad v_t(x,0) = -\sin 2\pi x
\end{cases}
$$

где:
- $\tilde{f}(x,t) = -\sin\pi x + \cos t - \sin t$ — неоднородность
- $\tilde{\theta}(x) = 3\sin\pi x$ — начальное смещение
- $\tilde{\psi}(x) = -\sin 2\pi x$ — начальная скорость

**Важно:** В отличие от параболических уравнений, здесь у нас **два** начальных условия (смещение и скорость).


## Шаг 3: Определение собственных функций и собственных значений

### Разделение переменных

Рассматриваем однородную задачу (без неоднородности):
$$v_{tt} = v_{xx}, \quad v(0,t) = 0, \quad v(1,t) = 0$$

Ищем решение в виде $v(x,t) = X(x) \cdot T(t)$.

Подставляем в уравнение:
$$X(x) \cdot T''(t) = X''(x) \cdot T(t)$$

Разделяя переменные:
$$\frac{T''(t)}{T(t)} = \frac{X''(x)}{X(x)} = -\lambda^2$$

### Пространственное уравнение

Получаем задачу Штурма-Лиувилля:
$$X''(x) = -\lambda^2 X(x)$$

Или в стандартном виде:
$$X''(x) + \lambda^2 X(x) = 0$$

**Граничные условия:** $X(0) = 0$ и $X(1) = 0$ (условия Дирихле)

### Общее решение пространственного уравнения

**Общее решение:**
$$X(x) = c_1 \sin\lambda x + c_2 \cos\lambda x$$


### Применение граничных условий

**Первое условие:** $X(0) = 0$

Подставляем $x = 0$:
$$X(0) = c_1 \sin(0) + c_2 \cos(0) = 0 + c_2 \cdot 1 = c_2 = 0$$

Следовательно: $X(x) = c_1 \sin\lambda x$

**Второе условие:** $X(1) = 0$

Подставляем $x = 1$:
$$X(1) = c_1 \sin\lambda = 0$$

Для нетривиального решения ($c_1 \neq 0$): $\sin\lambda = 0$

### Нахождение собственных значений

Уравнение $\sin\lambda = 0$ имеет решения:
$$\lambda = k\pi, \quad k = 1, 2, 3, \ldots$$

**Собственные значения:**
$$\lambda_k = k\pi, \quad k = 1, 2, 3, \ldots$$

### Собственные функции

Для каждого собственного значения $\lambda_k$ получаем соответствующую собственную функцию:
$$X_k(x) = \sin(k\pi x), \quad k = 1, 2, 3, \ldots$$

### Итог шага 3

Мы нашли:
- **Собственные значения:** $\lambda_k = k\pi$, $k = 1, 2, 3, \ldots$
- **Собственные функции:** $X_k(x) = \sin(k\pi x)$, $k = 1, 2, 3, \ldots$

Каждая функция $X_k(x)$ удовлетворяет:
- Уравнению $X_k''(x) + \lambda_k^2 X_k(x) = 0$
- Граничным условиям $X_k(0) = 0$ и $X_k(1) = 0$


## Шаг 4: Разложение функций в ряды Фурье

### Представление решения в виде ряда

Ищем решение в виде бесконечного ряда по собственным функциям:
$$v(x,t) = \sum_{k=1}^{\infty} X_k(x) T_k(t) = \sum_{k=1}^{\infty} \sin(k\pi x) T_k(t)$$

### Разложение начального смещения

Разложим функцию $\tilde{\theta}(x) = 3\sin\pi x$ в ряд Фурье по системе собственных функций $\{\sin(k\pi x)\}$.

**Важное наблюдение:** Так как $3\sin\pi x$ уже является собственной функцией при $k=1$ ($X_1(x) = \sin(\pi x)$), то:

- $\theta_1 = 3$ (наша функция полностью совпадает с первой гармоникой, умноженной на 3)
- $\theta_k = 0$ для всех $k \neq 1$ (в начальном смещении нет других гармоник)

### Разложение начальной скорости

Разложим функцию $\tilde{\psi}(x) = -\sin 2\pi x$ в ряд Фурье.

**Важное наблюдение:** Так как $-\sin 2\pi x$ уже является собственной функцией при $k=2$ ($X_2(x) = \sin(2\pi x)$), то:

- $\psi_2 = -1$ (наша функция полностью совпадает со второй гармоникой, умноженной на $-1$)
- $\psi_k = 0$ для всех $k \neq 2$ (в начальной скорости нет других гармоник)


### Разложение неоднородности

Теперь разложим неоднородность $\tilde{f}(x,t) = -\sin\pi x + \cos t - \sin t$:

$$\tilde{f}(x,t) = \sum_{k=1}^{\infty} \mu_k(t) X_k(x) = \sum_{k=1}^{\infty} \mu_k(t) \sin(k\pi x)$$

**Коэффициенты Фурье** вычисляются по формуле:
$$\mu_k(t) = 2\int_0^1 [-\sin\pi x + \cos t - \sin t]\sin(k\pi x) dx$$

Разделяем интеграл:
$$= 2\int_0^1 (-\sin\pi x)\sin(k\pi x) dx + 2(\cos t - \sin t)\int_0^1 \sin(k\pi x) dx$$

**Вычисляем первый интеграл:**
- При $k = 1$: $\int_0^1 \sin^2(\pi x) dx = \frac{1}{2}$, поэтому $-2 \cdot \frac{1}{2} = -1$
- При $k \neq 1$: интеграл равен нулю (ортогональность)

**Вычисляем второй интеграл:**
$$\int_0^1 \sin(k\pi x) dx = \frac{1 - \cos(k\pi)}{k\pi}$$

Так как $\cos(k\pi) = (-1)^k$:
- При нечетных $k$: интеграл равен $\frac{2}{k\pi}$
- При четных $k$: интеграл равен $0$


### Итоговые коэффициенты

**При $k = 1$:**
$$\mu_1(t) = -1 + 2(\cos t - \sin t) \cdot \frac{2}{\pi} = -1 + \frac{4(\cos t - \sin t)}{\pi}$$

**При нечетных $k \geq 3$:**
$$\mu_k(t) = 0 + 2(\cos t - \sin t) \cdot \frac{2}{k\pi} = \frac{4(\cos t - \sin t)}{k\pi}$$

**При четных $k$:**
$$\mu_k(t) = 0$$ (оба интеграла равны нулю)

### Итог шага 4

**Разложение начального смещения:**
$$3\sin\pi x = \sum_{k=1}^{\infty} \theta_k \sin(k\pi x) = \theta_1 \sin(\pi x) = 3 \sin(\pi x)$$

где $\theta_1 = 3$, $\theta_k = 0$ для $k \neq 1$.

**Разложение начальной скорости:**
$$-\sin 2\pi x = \sum_{k=1}^{\infty} \psi_k \sin(k\pi x) = \psi_2 \sin(2\pi x) = -\sin(2\pi x)$$

где $\psi_2 = -1$, $\psi_k = 0$ для $k \neq 2$.

**Разложение неоднородности:**
$$\tilde{f}(x,t) = \sum_{k=1}^{\infty} \mu_k(t) \sin(k\pi x)$$

где $\mu_k(t)$ вычислены выше и зависят от времени через $\cos t$ и $\sin t$.


## Шаг 5: Решение уравнений для $T_k(t)$

### Подстановка ряда в уравнение

Подставляем ряд $v(x,t) = \sum_{k=1}^{\infty} \sin(k\pi x) T_k(t)$ в уравнение $v_{tt} = v_{xx} + \tilde{f}(x,t)$:

**Левая часть:**
$$\frac{\partial^2}{\partial t^2}\left[\sum_{k=1}^{\infty} \sin(k\pi x) T_k(t)\right] = \sum_{k=1}^{\infty} \sin(k\pi x) T_k''(t)$$

**Правая часть:**
$$\frac{\partial^2}{\partial x^2}\left[\sum_{k=1}^{\infty} \sin(k\pi x) T_k(t)\right] + \tilde{f}(x,t)$$

Используя $X_k''(x) = -\lambda_k^2 X_k(x) = -\lambda_k^2 \sin(k\pi x)$:
$$= \sum_{k=1}^{\infty} (-\lambda_k^2) \sin(k\pi x) T_k(t) + \sum_{k=1}^{\infty} \mu_k(t) \sin(k\pi x)$$

### Приравнивание коэффициентов

Приравниваем коэффициенты при одинаковых функциях $\sin(k\pi x)$:

$$\sum_{k=1}^{\infty} \sin(k\pi x) T_k''(t) = \sum_{k=1}^{\infty} (-\lambda_k^2) \sin(k\pi x) T_k(t) + \sum_{k=1}^{\infty} \mu_k(t) \sin(k\pi x)$$

Отсюда для каждого $k$:
$$T_k''(t) = -\lambda_k^2 T_k(t) + \mu_k(t), \quad k = 1, 2, \ldots$$

**Начальные условия:** $T_k(0) = \theta_k$, $T_k'(0) = \psi_k$

**Важно:** В отличие от параболических уравнений, здесь уравнение **второго порядка** по времени, поэтому нужны два начальных условия.


### Решение неоднородного уравнения методом вариации постоянных

Это линейное неоднородное ОДУ второго порядка:
$$T_k''(t) + \lambda_k^2 T_k(t) = \mu_k(t)$$

**Общее решение однородного уравнения:**
$$T_{k\text{одн}}(t) = C_1 \cos(\lambda_k t) + C_2 \sin(\lambda_k t)$$

**Решение неоднородного уравнения** методом вариации постоянных:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

### Физический смысл

Решение состоит из трех частей:
1. **Свободные колебания** от начального смещения: $\theta_k\cos(\lambda_k t)$
2. **Свободные колебания** от начальной скорости: $\frac{\psi_k}{\lambda_k}\sin(\lambda_k t)$
3. **Вынужденные колебания** от источника: $\frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$

### Для нашей задачи

**При $k = 1$:**
- $\lambda_1 = \pi$, $\theta_1 = 3$, $\psi_1 = 0$
- $\mu_1(t) = -1 + \frac{4(\cos t - \sin t)}{\pi}$

**При $k = 2$:**
- $\lambda_2 = 2\pi$, $\theta_2 = 0$, $\psi_2 = -1$
- $\mu_2(t) = 0$

**При $k \geq 3$:**
- $\theta_k = 0$, $\psi_k = 0$
- $\mu_k(t)$ ненулевые только для нечетных $k$

### Итог шага 5

Решение временных уравнений:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

где $\lambda_k = k\pi$, а $\theta_k$, $\psi_k$ и $\mu_k(t)$ найдены на шаге 4.


## Шаг 6: Запись полного решения для $v(x,t)$

### Общая формула

На шаге 4 мы записали решение в виде ряда:
$$v(x,t) = \sum_{k=1}^{\infty} \sin(k\pi x) T_k(t)$$

Подставляем найденные выражения для $T_k(t)$ из шага 5:
$$v(x,t) = \sum_{k=1}^{\infty} \sin(k\pi x) \left[\theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau\right]$$

### Главные члены

Так как $\theta_1 = 3$, $\psi_2 = -1$, а остальные $\theta_k$ и $\psi_k$ равны нулю, главные вклады:
- От начального смещения: $3\sin(\pi x)\cos(\pi t)$
- От начальной скорости: $-\frac{1}{2\pi}\sin(2\pi x)\sin(2\pi t)$
- От неоднородности: интегральные члены от $\mu_k(t)$

### Итог шага 6

Полное решение для функции $v(x,t)$:
$$v(x,t) = \sum_{k=1}^{\infty} \sin(k\pi x) T_k(t)$$

где $T_k(t)$ находятся из уравнений на шаге 5 и содержат осцилляции (колебания) с частотами $\lambda_k = k\pi$.


## Шаг 7: Окончательный ответ

### Возврат к исходной функции

На шаге 1 мы установили: $u(x,t) = v(x,t) + s(x,t)$

Подставляем:
- $v(x,t)$ из шага 6
- $s(x,t) = (1-x)\cos t + x\sin t$ из шага 1

### Финальный ответ

**Ответ задачи №5:**

$$u(x,t) = v(x,t) + s(x,t) = \sum_{k=1}^{\infty} T_k(t)\sin(k\pi x) + (1-x)\cos t + x\sin t$$

где:
$$T_k(t) = \theta_k\cos(\lambda_k t) + \frac{\psi_k}{\lambda_k}\sin(\lambda_k t) + \frac{1}{\lambda_k}\int_0^t \mu_k(\tau)\sin[\lambda_k(t-\tau)] d\tau$$

$$\lambda_k = k\pi$$

$$\theta_k = \begin{cases} 3, & k=1 \\ 0, & k \neq 1 \end{cases}$$

$$\psi_k = \begin{cases} -1, & k=2 \\ 0, & k \neq 2 \end{cases}$$

а коэффициенты $\mu_k(t)$ вычислены на шаге 4 и зависят от времени через $\cos t$ и $\sin t$.

### Физическая интерпретация решения

Решение состоит из трех компонент:
1. **Свободные колебания струны** с собственными частотами $\lambda_k = k\pi$ (гармоники)
2. **Вынужденные колебания** от источника $f(x,t)$ (интегральные члены)
3. **Вклад от неоднородных граничных условий:** $(1-x)\cos t + x\sin t$ — границы колеблются во времени

**Важно:** В отличие от параболических уравнений, решение не затухает, а продолжает колебаться со временем (это характерно для волновых процессов).


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import odeint

# Параметры
x = np.linspace(0, 1, 100)
t_values = [0, 0.5, 1.0, 1.5, 2.0]
t_surf = np.linspace(0, 2, 50)
N_TERMS = 10

# Функция s(x,t)
def s_5(x, t):
    return (1 - x) * np.cos(t) + x * np.sin(t)

# Функция mu_k(t)
def mu_k_5(t, k):
    if k % 2 == 0:  # Четные k
        return 0
    elif k == 1:
        return -1 + (4 / np.pi) * (np.cos(t) - np.sin(t))
    else:  # Нечетные k != 1
        return (4 / (k * np.pi)) * (np.cos(t) - np.sin(t))

# Модель ОДУ для T_k(t)
def model_5(Y, t, k):
    T, T_prime = Y
    lam_k = k * np.pi
    dYdt = [T_prime, mu_k_5(t, k) - lam_k**2 * T]
    return dYdt

# Решение ОДУ для T_k(t)
def solve_T_k_5(k, t_points):
    theta_k = 3 if k == 1 else 0
    psi_k = -1 if k == 2 else 0
    Y0 = [theta_k, psi_k]
    sol = odeint(model_5, Y0, t_points, args=(k,))
    return sol[:, 0]

# Вычисление u(x,t)
def u_5_analytical(x, t, T_solutions):
    s_xt = s_5(x, t)
    v_xt = sum(T_k * np.sin((k+1) * np.pi * x) for k, T_k in enumerate(T_solutions))
    return v_xt + s_xt

# Решение ОДУ для всех k
T_solutions_5 = [solve_T_k_5(k, t_surf) for k in range(1, N_TERMS + 1)]

# Создание фигуры
fig = plt.figure(figsize=(16, 6))

# График 1: Зависимость от x при разных t
ax1 = fig.add_subplot(121)
for t_val in t_values:
    t_idx = np.abs(t_surf - t_val).argmin()
    T_snapshot = [T_sol[t_idx] for T_sol in T_solutions_5]
    u = u_5_analytical(x, t_surf[t_idx], T_snapshot)
    ax1.plot(x, u, label=f'$t = {t_val:.1f}$')
ax1.set_xlabel('$x$', fontsize=12)
ax1.set_ylabel('$u(x,t)$', fontsize=12)
ax1.set_title(f'Задача №5: $u(x,t)$ при различных $t$ (N={N_TERMS})', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim([0, 1])

# График 2: 3D поверхность u(x,t)
ax2 = fig.add_subplot(122, projection='3d')
X, T = np.meshgrid(x, t_surf)
U = np.array([[u_5_analytical(xi, t_surf[i], [T_sol[i] for T_sol in T_solutions_5]) for xi in x] for i in range(len(t_surf))])

surf = ax2.plot_surface(X, T, U, cmap='viridis', alpha=0.9, linewidth=0, antialiased=True)
ax2.set_xlabel('$x$', fontsize=12)
ax2.set_ylabel('$t$', fontsize=12)
ax2.set_zlabel('$u(x,t)$', fontsize=12)
ax2.set_title(f'Задача №5: Поверхность $u(x,t)$ (N={N_TERMS})', fontsize=14)
fig.colorbar(surf, ax=ax2, shrink=0.5, aspect=5, label='$u(x,t)$')

plt.tight_layout()
plt.show()
print("Визуализация задачи №5")
```
