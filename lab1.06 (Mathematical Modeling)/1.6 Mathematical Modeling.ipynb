```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

T0 = 100.0   # Начальная температура
T_env = 22.0 # Температура окружающей среды
t_end = 60.0 # Время в минутах
dt = 0.5 

k_values = [0.05, 0.1, 0.2]
colors = ['red', 'orange', 'green']
```

```python
def cooling_model(T, T_env, k):
    return -k * (T - T_env)

def solve_cooling(T0, T_env, k, t_end, dt):
    t = np.arange(0, t_end + dt, dt)
    T = np.zeros_like(t)
    T[0] = T0
    for i in range(len(t)-1):
        T[i+1] = T[i] + dt * cooling_model(T[i], T_env, k)
    return t, T
```

```python
plt.figure(figsize=(10, 6))

for k, color in zip(k_values, colors):
    t, T = solve_cooling(T0, T_env, k, t_end, dt)
    plt.plot(t, T, color=color, linewidth=2.5, label=f'k = {k} (Скорость остывания)')

plt.axhline(T_env, color='blue', linestyle='--', linewidth=2, label='Температура среды (22°C)')

plt.title('Модель закона охлаждения Ньютона')
plt.xlabel('Время (мин)')
plt.ylabel('Температура (°C)')
plt.legend()
plt.tight_layout()
plt.show()
```
