# Подробный разбор решения Задач №9-10 (Эллиптические уравнения - Полярные координаты)

## Постановка задач

**Уравнение:** $\Delta u = 0$ в полярных координатах $(r, \theta)$

**Задача №9 (внутренняя):** $r \leq 1$
- Граничное условие: $u(1,\theta) = -\cos 3\theta - \sin 4\theta$

**Задача №10 (внешняя):** $r \geq 1$
- $u(1,\theta) = 3 - 2\cos 3\theta$
- $u(3,\theta) = 4 + \cos 6\theta$

## Физический смысл и особенности

**Полярные координаты** используются, когда область имеет круговую симметрию: круг, кольцо, сектор.

Оператор Лапласа в полярных координатах:
$$\Delta u = \frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial u}{\partial r}\right) + \frac{1}{r^2}\frac{\partial^2 u}{\partial\theta^2}$$

**Ключевые моменты:**

1. **Условие периодичности:** $u(r, \theta + 2\pi) = u(r, \theta)$ приводит к дискретному спектру по углу: $k = 0, 1, 2, \ldots$

2. **Уравнение Эйлера для $R(r)$:** $r^2R'' + rR' - k^2R = 0$ имеет решения вида $R_k(r) = A_k r^k + B_k r^{-k}$

3. **Условия регулярности:**
   - **Внутренняя задача:** $B_k = 0$ (иначе $r^{-k} \to \infty$ при $r \to 0$)
   - **Внешняя задача:** $A_k = 0$ (иначе $r^k \to \infty$ при $r \to \infty$)

4. **Простота решения:** Если граничные условия уже разложены в ряд Фурье (как в задаче №9), решение находится сразу!
