# Подробный разбор решения Задач №11-12 (Эллиптические уравнения - Полярные координаты с углом α)

## Постановка задач

**Уравнение:** $\Delta u = 0$ в полярных координатах $(r, \theta)$

**Область:** сектор круга $0 \leq \theta \leq \alpha = \frac{\pi}{3}$, $r \in [a, b]$

**Граничные условия:**
- $u_\theta(r, 0) = 0$ — изолированная радиальная граница при $\theta = 0$
- $u_\theta(r, \alpha) = 0$ — изолированная радиальная граница при $\theta = \alpha$
- Дополнительные условия на $r = a$ и $r = b$

## Физический смысл

Это задача для **сектора круга** (например, кусок пиццы). Условия $u_\theta = 0$ означают, что радиальные границы **теплоизолированы** — нет теплового потока через эти границы.

## Особенности решения

**Ключевое отличие от задач №9-10:**

Вместо условия периодичности $u(r, \theta + 2\pi) = u(r, \theta)$ имеем условия Неймана на радиальных границах: $u_\theta(r, 0) = u_\theta(r, \alpha) = 0$.

Это приводит к **другой системе собственных функций** по углу:
$$F_k(\theta) = \cos\frac{k\pi\theta}{\alpha}, \quad k = 0, 1, 2, \ldots$$

**Собственные значения:** $\lambda_k = \frac{k\pi}{\alpha}$

Для $R(r)$ получаем уравнение Эйлера с параметром $\lambda_k$:
$$r^2R'' + rR' - \lambda_k^2 R = 0$$

Решение: $R_k(r) = A_k r^{\lambda_k} + B_k r^{-\lambda_k}$

**Важно:** При $\alpha = \frac{\pi}{3}$ собственные значения равны $\lambda_k = 3k$, что приводит к целым степеням $r$ в решении.
