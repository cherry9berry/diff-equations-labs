## Задача №8 (Декартовы координаты)

**Дано:**
- **Уравнение:** $\Delta u = u_{xx} + u_{yy} = 0$
- **Область:** Прямоугольник $x \in [0,2], y \in [0,1]$
- **Граничные условия (Г.У.):**
  - $u_x(0,y) = 0$
  - $u_y(x,0) = 0$
  - $u_x(2,y) = 2\sin(3\pi y)$
  - $u_y(x,1) = 3\sin(\pi x)$

### Принцип суперпозиции
$u(x,y) = u_1(x,y) + u_2(x,y)$, где $u_1$ и $u_2$ решают следующие подзадачи.

---
### Подзадача для $u_1(x,y)$ (неоднородность на $y=1$)

$$
\begin{cases}
\Delta u_1 = 0 \\
u_{1,x}(0,y) = 0, \quad u_{1,x}(2,y) = 0 \\
u_{1,y}(x,0) = 0 \\
u_{1,y}(x,1) = 3\sin(\pi x)
\end{cases}
$$
Решение ищется в виде $u_1 = X(x)Y(y)$. **Задача Штурма-Лиувилля для $X(x)$:**
- $X'' + \lambda^2 X = 0$ с Г.У. $X'(0)=0, X'(2)=0$.
- **Собственные функции:** $X_k(x) = \cos\left(\frac{k\pi x}{2}\right)$, $\lambda_k = \frac{k\pi}{2}$, $k=0,1,2,...$
**Уравнение для $Y(y)$:** $Y'' - \lambda_k^2 Y = 0$ с Г.У. $Y'(0)=0$.
- **Решение:** $Y_k(y) = A_k \cosh(\lambda_k y)$ для $k \ge 1$, $Y_0(y)=B_0$ для $k=0$.
Общее решение для $u_1$: $u_1 = B_0 + \sum_{k=1}^\infty A_k \cos(\lambda_k x) \cosh(\lambda_k y)$.
Из $u_{1,y}(x,1) = 3\sin(\pi x)$ находим $A_k$:
$$ \sum_{k=1}^\infty A_k \lambda_k \sinh(\lambda_k) \cos(\lambda_k x) = 3\sin(\pi x) $$
$$ A_k \lambda_k \sinh(\lambda_k) = \frac{2}{2}\int_0^2 3\sin(\pi x)\cos(\lambda_k x)dx $$

---
### Подзадача для $u_2(x,y)$ (неоднородность на $x=2$)

$$
\begin{cases}
\Delta u_2 = 0 \\
u_{2,x}(0,y) = 0 \\
u_{2,y}(x,0) = 0, \quad u_{2,y}(x,1) = 0 \\
u_{2,x}(2,y) = 2\sin(3\pi y)
\end{cases}
$$
Решение ищется в виде $u_2 = X(x)Y(y)$. **Задача Штурма-Лиувилля для $Y(y)$:**
- $Y'' + \lambda^2 Y = 0$ с Г.У. $Y'(0)=0, Y'(1)=0$.
- **Собственные функции:** $Y_k(y) = \cos(k\pi y)$, $\lambda_k = k\pi$, $k=0,1,2,...$
**Уравнение для $X(x)$:** $X'' - \lambda_k^2 X = 0$ с Г.У. $X'(0)=0$.
- **Решение:** $X_k(x) = C_k \cosh(\lambda_k x)$ для $k \ge 1$, $X_0(x)=D_0$ для $k=0$.
Общее решение для $u_2$: $u_2 = D_0 + \sum_{k=1}^\infty C_k \cosh(\lambda_k x) \cos(\lambda_k y)$.
Из $u_{2,x}(2,y) = 2\sin(3\pi y)$ находим $C_k$:
$$ \sum_{k=1}^\infty C_k \lambda_k \sinh(2\lambda_k) \cos(\lambda_k y) = 2\sin(3\pi y) $$
$$ C_k \lambda_k \sinh(2\lambda_k) = 2\int_0^1 2\sin(3\pi y)\cos(\lambda_k y)dy $$


```python
# Задача №8: Визуализация
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad

# Параметры
N_TERMS = 50
x_points = np.linspace(0, 2, 50)
y_points = np.linspace(0, 1, 50)
X, Y = np.meshgrid(x_points, y_points)

# Коэффициенты A_k для u1
A_k = np.zeros(N_TERMS + 1)
for k in range(1, N_TERMS + 1):
    lam_k = k * np.pi / 2
    integrand = lambda x: 3 * np.sin(np.pi * x) * np.cos(lam_k * x)
    integral, _ = quad(integrand, 0, 2)
    A_k[k] = integral / (lam_k * np.sinh(lam_k)) if lam_k > 0 and np.sinh(lam_k) > 0 else 0

# Коэффициенты C_k для u2
C_k = np.zeros(N_TERMS + 1)
for k in range(1, N_TERMS + 1):
    lam_k = k * np.pi
    integrand = lambda y: 2 * np.sin(3 * np.pi * y) * np.cos(lam_k * y)
    integral, _ = quad(integrand, 0, 1)
    C_k[k] = (2 * integral) / (lam_k * np.sinh(2 * lam_k)) if lam_k > 0 and np.sinh(2*lam_k) > 0 else 0
    
# Вычисление u1, u2 и u
U1 = sum(A_k[k] * np.cos(k*np.pi*X/2) * np.cosh(k*np.pi*Y/2) for k in range(1,N_TERMS+1))
U2 = sum(C_k[k] * np.cosh(k*np.pi*X) * np.cos(k*np.pi*Y) for k in range(1,N_TERMS+1))
U = U1 + U2

# Визуализация
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, U, cmap='viridis', edgecolor='none')
ax.set_title(f'Решение $u(x,y)$ (N={N_TERMS})', fontsize=16)
ax.set_xlabel('$x$'); ax.set_ylabel('$y$'); ax.set_zlabel('$u(x,y)$')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
