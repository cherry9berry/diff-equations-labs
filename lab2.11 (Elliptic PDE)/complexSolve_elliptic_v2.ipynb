## Задача №8 (Декартовы координаты)

**Дано:**
- **Уравнение:** $\Delta u = u_{xx} + u_{yy} = 0$
- **Область:** Прямоугольник $x \in [0,2], y \in [0,1]$
- **Граничные условия:**
  - $u_x(0,y) = 0$
  - $u_y(x,0) = 0$
  - $u_x(2,y) = 2\sin 3\pi y$
  - $u_y(x,1) = 3\sin\pi x$

### Решение через принцип суперпозиции

Представим $u(x,y) = u_1(x,y) + u_2(x,y)$, где:

**Подзадача 1:** $\Delta u_1 = 0$, $u_{1,x}(0,y) = 0$, $u_{1,x}(2,y) = 0$, $u_{1,y}(x,0) = 0$, $u_{1,y}(x,1) = 3\sin\pi x$

**Подзадача 2:** $\Delta u_2 = 0$, $u_{2,x}(0,y) = 0$, $u_{2,y}(x,0) = 0$, $u_{2,y}(x,1) = 0$, $u_{2,x}(2,y) = 2\sin 3\pi y$

### Подзадача 1:
Собственные функции по $x$: $X_k(x) = \cos\frac{k\pi x}{2}$, $\lambda_k = \frac{k\pi}{2}$

Решение: $u_1 = \sum_{k=1}^{\infty} A_k \cos\frac{k\pi x}{2} \cosh\frac{k\pi y}{2}$

Из Г.У. при $y=1$: $A_k \frac{k\pi}{2} \sinh\frac{k\pi}{2} = \int_0^2 3\sin\pi x \cos\frac{k\pi x}{2}dx$

### Подзадача 2:
Собственные функции по $y$: $Y_k(y) = \cos(k\pi y)$, $\lambda_k = k\pi$

Решение: $u_2 = \sum_{k=1}^{\infty} C_k \cosh(k\pi x) \cos(k\pi y)$

Из Г.У. при $x=2$: $C_k k\pi \sinh(2k\pi) = 2\int_0^1 2\sin 3\pi y \cos(k\pi y)dy$

### Окончательное решение:

$$\boxed{u(x,y) = u_1(x,y) + u_2(x,y)}$$


```python
# Задача №8: Визуализация
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad

N = 50
x = np.linspace(0, 2, 50)
y = np.linspace(0, 1, 50)
X, Y = np.meshgrid(x, y)

# Вычисление коэффициентов A_k для u1
A_k = np.zeros(N+1)
for k in range(1, N+1):
    lam_k = k * np.pi / 2
    integrand = lambda x: 3 * np.sin(np.pi * x) * np.cos(lam_k * x)
    integral, _ = quad(integrand, 0, 2)
    A_k[k] = integral / (lam_k * np.sinh(lam_k)) if np.sinh(lam_k) > 0 else 0

# Вычисление коэффициентов C_k для u2
C_k = np.zeros(N+1)
for k in range(1, N+1):
    lam_k = k * np.pi
    integrand = lambda y: 2 * np.sin(3 * np.pi * y) * np.cos(lam_k * y)
    integral, _ = quad(integrand, 0, 1)
    C_k[k] = (2 * integral) / (lam_k * np.sinh(2 * lam_k)) if np.sinh(2*lam_k) > 0 else 0

# Вычисление u(x,y)
U1 = sum(A_k[k] * np.cos(k*np.pi*X/2) * np.cosh(k*np.pi*Y/2) for k in range(1,N+1))
U2 = sum(C_k[k] * np.cosh(k*np.pi*X) * np.cos(k*np.pi*Y) for k in range(1,N+1))
U = U1 + U2

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, U, cmap='viridis')
ax.set_title(f'Задача №8: u(x,y) (N={N})')
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u')
fig.colorbar(surf, shrink=0.5)
plt.show()
print("Визуализация задачи №8")
```

## Задача №9 (Полярные координаты, внутренняя задача)

**Дано:**
- **Уравнение:** $\Delta u = u_{rr} + \frac{1}{r}u_r + \frac{1}{r^2}u_{\theta\theta} = 0$
- **Область:** Круг $r \leq 1$
- **Граничное условие:** $u(1,\theta) = -\cos 3\theta - \sin 4\theta$

### Решение:

Для внутренней задачи Дирихле общее решение:
$$u(r,\theta) = A_0 + \sum_{k=1}^{\infty} r^k (A_k \cos k\theta + B_k \sin k\theta)$$

Из Г.У. на $r=1$:
$$u(1,\theta) = A_0 + \sum_{k=1}^{\infty} (A_k \cos k\theta + B_k \sin k\theta) = -\cos 3\theta - \sin 4\theta$$

Сравнивая коэффициенты: $A_0 = 0$, $A_3 = -1$, $B_4 = -1$, остальные $= 0$

### Окончательное решение:

$$\boxed{u(r,\theta) = -r^3 \cos 3\theta - r^4 \sin 4\theta}$$


```python
# Задача №9: Визуализация
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

r = np.linspace(0, 1, 50)
theta = np.linspace(0, 2*np.pi, 100)
R, THETA = np.meshgrid(r, theta)

# Решение u(r,theta)
U = -R**3 * np.cos(3*THETA) - R**4 * np.sin(4*THETA)

# Преобразование в декартовы координаты для визуализации
X = R * np.cos(THETA)
Y = R * np.sin(THETA)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, U, cmap='viridis')
ax.set_title('Задача №9: u(r,θ)')
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u')
fig.colorbar(surf, shrink=0.5)
plt.show()
print("Визуализация задачи №9")
```

## Задача №10 (Полярные координаты, кольцо)

**Дано:**
- **Область:** Кольцо $1 \leq r \leq 3$
- **Граничные условия:** $u(1,\theta) = 3 - 2\cos 3\theta$, $u(3,\theta) = 4 + \cos 6\theta$

### Решение:

Общее решение в кольце:
$$u(r,\theta) = E + G\ln r + \sum_{k=1}^{\infty} [(A_k r^k + B_k r^{-k})\cos k\theta + (C_k r^k + D_k r^{-k})\sin k\theta]$$

Из Г.У. составляется система для коэффициентов. Решение находится численно.
